# Train and test data

```python
x = df.drop('Survived', axis=1).values
y = df['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.3, random_state =1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver = 'liblinear')
```

```python
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(ytest, predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest, predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest, predictions)
```

# SUV Data Analysis

```python
suv_data = pd.read_csv('data/suv_data.csv')
```

```python
suv_data.head()
```

```python
print('# of entries is ' + str(len(suv_data)))
```

```python
male = pd.get_dummies(suv_data['Gender'], drop_first=True)
```

```python
x = pd.concat([suv_data.loc[:,['Age','EstimatedSalary']], male], axis=1)
```

```python
x
```

```python
suv_data
```

```python
x = x.values
y = suv_data['Purchased'].values
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
model = LogisticRegression(solver = 'liblinear', random_state=0)
```

```python
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
accuracy_score(ytest, predictions)
```
